# Single Perceptron for NOR Logic Gate for Bipolar Inputs

```python
def ActivationFunction(x):
    if x >= 1:
        return 1
    else:
        return -1
```

```python
def Perceptron(x1, x2, target):
    # Initialize weights and bias
    w1 = 0.0
    w2 = 0.0
    b = 0.0
    learning_rate = 0.3
    
    # Training the perceptron
    for _ in range(10):
        # Calculate the weighted sum
        weighted_sum = (w1 * x1) + (w2 * x2) + b
        
        # Apply activation function
        output = ActivationFunction(weighted_sum)
        
        # Update weights and bias based on error
        error = target - output
        w1 += learning_rate * error * x1
        w2 += learning_rate * error * x2
        b += learning_rate * error
        
    return w1, w2, b, weighted_sum
```

```python
# Input data for NOR gate (bipolar inputs)

X = ([1,1],[1,-1], [-1,1], [-1,-1])
Y = [-1, -1, -1, 1] # Target outputs for NOR gate

for i in range(len(X)):
    x1, x2 = X[i]
    target = Y[i]
    w1, w2, b, weighted_sum = Perceptron(x1, x2, target)
    print(f"Input: {x1}, {x2} => Weights: {int(w1)}, {int(w2)}, Bias: {int(b)}, Output: {int(weighted_sum)}")
```
