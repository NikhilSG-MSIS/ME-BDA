```python
import mysql.connector
import csv

connection = mysql.connector.connect(
    host='localhost',
    user='root',
    password='Msis@123',
    database='financial_entities_db'
)

cursor = connection.cursor()

if connection.is_connected():
        print("Successfully connected to MySQL database")

        cursor = connection.cursor()

        # Create table without sr_no
        create_table_query = '''
        CREATE TABLE IF NOT EXISTS entities (
            entity_name VARCHAR(100),
            category VARCHAR(50),
            regulator VARCHAR(20),
            fip_implementation_stage VARCHAR(20),
            fiu_implementation_stage VARCHAR(20)
        );
        '''

        cursor.execute(create_table_query)
        print("Table 'entities' created successfully (or already exists)")

with open('Sahamati.csv', 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    next(reader)  # Skip header

    for row in reader:
        data = tuple(row[1:])  # Skip Sr. No
        cursor.execute(
            "INSERT INTO entities (entity_name, category, regulator, fip_implementation_stage, fiu_implementation_stage) VALUES (%s, %s, %s, %s, %s)",
            data
        )

connection.commit()
cursor.close()
connection.close()
print("Data inserted successfully.")
```

```python
import mysql.connector

# Connect to MySQL
connection = mysql.connector.connect(
    host='localhost',
    user='root',
    password='Msis@123',
    database='financial_entities_db'
)

cursor = connection.cursor()

# Execute SQL query to count entities for each regulator
cursor.execute("""
    SELECT regulator, COUNT(*) AS entity_count
    FROM entities
    GROUP BY regulator;
""")

# Fetch all results
results = cursor.fetchall()

# Print results
for row in results:
    print(f"Regulator: {row[0]}, Entity Count: {row[1]}")

# Clean up
cursor.close()
connection.close()
```
